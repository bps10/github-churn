```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from github3 import login

import time
import helper as h

gh = login('bps10', password='boniver#3')
```

```python
df = pd.read_csv('events_data/events_2016_01_01_2016_06_01.csv')
df = df[df.event_count > 1]
df['last_event'] = pd.to_datetime(df.last_event)
df['first_event'] = pd.to_datetime(df.first_event)
df['time_between_first_last_event'] = df.last_event - df.first_event
df.head()
```

```python
existing_users = pd.read_csv('all_users.csv')
existing_users = set(existing_users.login.values.tolist())
```

```python
random_indexes = np.arange(len(df))
np.random.seed(0)
np.random.shuffle(random_indexes)
random_indexes[:4]
```

```python
start_index = 205000
for t in range(5):
    # record batch start time
    start_time = time.time()

    user_batch, end_index = h.get_batch(gh, df, random_indexes, start_index, existing_users,
                                        batch_size=5000)

    user_batch = pd.DataFrame.from_dict(user_batch, orient='index')
    print(user_batch.head())
    
    # save to file
    user_batch.to_csv('user_batch_{0}_{1}.csv'.format(start_index, end_index))
    start_index = end_index
    print('Starting index {0}.'.format(start_index))
    # record end of batch time
    end_time = time.time()

    # put the loop to sleep until 1 hour has elapsed since the start of the batch
    time.sleep(60*60 - (end_time - start_time) + 60)
    
print(end_index)
```
